```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Personal/human_activity_segmentation_project/data/raw/subjects_information.csv')
```

```python
df
```

```python
print("DataFrame Info:")
df.info()

print("\nDescriptive Statistics:")
df.describe()

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
categorical_cols = ['Sex', 'Dominant hand']

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

for ax, col in zip(axes, categorical_cols):
    sns.countplot(x=col, data=df, palette='viridis', ax=ax)

    ax.set_title(f"Distribution of {col}",
                 fontsize=16,
                 fontweight='bold',
                 pad=15,
                 color="#1a1a1a")

    ax.set_xlabel(col, fontsize=14)
    ax.set_ylabel("Count", fontsize=14)

    ax.grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
colors = ["#4C72B0", "#55A868", "#C44E52", "#8172B3", "#CCB974"]

numerical_cols = ['Age (years)', 'Height (cm)', 'Weight (kg)',
                  'Resting HR (bpm)', 'Max HR (bpm)']

for i, col in enumerate(numerical_cols):
    plt.figure(figsize=(12,5))

    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True, color=colors[i], alpha=0.7)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col], color=colors[i])

    mean_val = df[col].mean()
    plt.scatter(mean_val, 0,
                color="black",
                s=120,
                marker="X",
                zorder=10)

    plt.title(f"Box Plot of {col} (Mean marked)")
    plt.xlabel(col)

    plt.tight_layout()
    plt.show()
```
